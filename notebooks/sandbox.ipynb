```python
import matplotlib.pyplot as plt
import random
from src.data import OxfordPetsSegmentation
import torch
from torch.utils.data import DataLoader

num = random.randint(0, 1000)

dataset = OxfordPetsSegmentation(root="data/oxford_pets", split="train", transform=None, target_transform=None, seed=num)
random.shuffle(dataset.samples)
dataset_train = OxfordPetsSegmentation(root="data/oxford_pets", split="train", transform=None, target_transform=None, seed = num)
dataset_val = OxfordPetsSegmentation(root="data/oxford_pets", split="val", transform=None, target_transform=None, seed = num)
dataset_test = OxfordPetsSegmentation(root="data/oxford_pets", split="test", transform=None, target_transform=None, seed = num)

train_image, train_mask = dataset_train[0]
val_image, val_mask = dataset_val[0]
test_image, test_mask = dataset_test[0]

plt.subplot(3, 2, 1)
plt.imshow(train_image)
plt.subplot(3, 2, 2)
plt.imshow(train_mask)
plt.subplot(3, 2, 3)
plt.imshow(val_image)
plt.subplot(3, 2, 4)
plt.imshow(val_mask)
plt.subplot(3, 2, 5)
plt.imshow(test_image)
plt.subplot(3, 2, 6)
plt.imshow(test_mask)
```
